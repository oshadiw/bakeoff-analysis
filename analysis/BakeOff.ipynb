```python
import pandas as pd
import numpy as np
# Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup as soup
```

```python
# Set the executable path and initialize the chrome browser in splinter

executable_path = {'executable_path': 'chromedriver'}
browser = Browser('chrome', **executable_path)
```

```python
url = 'https://thegreatbritishbakeoff.co.uk/recipes/all/?filter[collection]=81&showFilter=false'
browser.visit(url)
browser.links.find_by_partial_text('OK').click()
```

```python
#Last page of the browser
last_url = browser.find_by_css('a.page-numbers')[-1].value
last_url
```

```python
# Creating an empty list to store the titles, image, servings
technical_list=[]

# Creating an empty dictionary
technical_dict ={'title':[], 'judge':[], 'makes':[], 'difficulty':[], 'hands_on_time':[], 'baking_time':[], 'ingredients':[], 'recipe_image':[], 'judge_image':[]}
```

```python
nRecipe = browser.find_by_css('div.recipes-loop__item.collection')
```

```python
def page(browser):

    for i in range(len(nRecipe)):
                        
                # loading the page
                browser.find_by_css("div.recipes-loop__item.collection", wait_time=1)[i].click()
                recipe_html = browser.html
                recipe_soup = soup(recipe_html, 'html.parser')
                               
                title = recipe_soup.find('h1').text
                h5=recipe_soup.find('h5')
                
                global j0
                try:
                    j0 = recipe_soup.h5.find('a').text
                except AttributeError:
                    print("judge not found")                    
                judge = j0
                
                global x0
                try:
                    x0 = recipe_soup.find_all('div', class_='recipe-details__item')[0]
                except UnboundLocalError:
                    print("not found")
                except IndexError:
                    print("not found")
                makes = x0.find('span', class_='value').get_text()

                global x1
                try:
                    x1 = recipe_soup.find_all('div', class_='recipe-details__item')[1]
                except UnboundLocalError:
                    print("not found")
                except IndexError:
                    print("not found")
                difficulty = x1.find('span', class_='value').get_text()
    
                global x2
                try:
                    x2 = recipe_soup.find_all('div', class_='recipe-details__item')[2]
                except UnboundLocalError:
                    print("not found")
                except IndexError:
                    print("not found")
                hands_on_time = x2.find('span', class_='value').get_text()
                
                global x3
                try:
                    x3 = recipe_soup.find_all('div', class_='recipe-details__item')[3]
                except UnboundLocalError:
                    print("not found")
                except IndexError:
                    print("not found")                
                baking_time = x3.find('span', class_='value').get_text()
                
                try:
                    i0 = recipe_soup.find('div', class_='recipe-sidebar__section__content').findAll()
#                    i0 = recipe_soup.find('div', class_='recipe-sidebar__section recipe-sidebar__section--ingredients').findAll('p')
                except AttributeError:
                    print("not found")
                ingredients = i0


                container = recipe_soup.find('div', class_='container')               
                recipe_image = container.find_all('img')[0].get('src')
                
                global j1
                try:
                    j1 = container.find_all('img')[1].get('src')
                except UnboundLocalError:
                    print("not found")
                except IndexError:
                    print("judge not found")                
                judge_image = j1
                    


                # Creating dictionary keys
                technical_dict['title'].append(title)
                technical_dict['judge'].append(judge)
                technical_dict['makes'].append(makes)
                technical_dict['difficulty'].append(difficulty)
                technical_dict['hands_on_time'].append(hands_on_time)
                technical_dict['baking_time'].append(baking_time)
                technical_dict['ingredients'].append(ingredients)
                technical_dict['recipe_image'].append(recipe_image)
                technical_dict['judge_image'].append(judge_image)

                technical_dictionary = {'title':title, 'judge':judge, 'makes':makes, 'difficulty':difficulty, 'hands_on_time':hands_on_time, 'baking_time':baking_time, 'ingredients':ingredients, 'recipe_image':recipe_image, 'judge_image':judge_image}

                technical_list.append(technical_dictionary)

                browser.back()
    return technical_list
```

```python
for j in range(1, int(last_url)+1):
    
    if j == 1:
        page(browser)
    else:
        link1 = 'https://thegreatbritishbakeoff.co.uk/recipes/all/page/'
        link2 = '/?filter%5Bcollection%5D=81&showFilter=false#038;showFilter=false'

        browser.links.find_by_href(link1+str(j)+link2).click()
        nRecipe = browser.find_by_css('div.recipes-loop__item.collection')

        page(browser)
```

```python
technical_list_df = pd.DataFrame(technical_list)
```
